```python

def bubble_sort(arr):

  for i in range(len(arr) - 1):
    flag = 0
    for j in range(len(arr) - 1 - i):
      if arr[j] > arr[j+1]:
        arr[j],arr[j+1] = arr[j+1],arr[j]
        flag =1

    if flag == 0:
      break
  
  return arr
```

```python
def selection_sort(arr): 
    for i in range(len(arr) - 1):
        min = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min]:
                min = j
        arr[i], arr[min] = arr[min], arr[i]

    print(arr)
```

```python
L = []

import random

for i in range(10000):
    L.append(random.randint(1, 10000))

len(L)
```

```python
L1 = L[:]
# Cloning
```

```python
import time

start = time.time()
bubble_sort(L)
print("Time taken",time.time() - start,"secs")
```

```python
import time

start = time.time()
selection_sort(L1)
print("Time taken",time.time() - start,"secs")
```

```python
# Selection sort is faster than bubble sort. because it makes fewer comparisons and swaps. Bubble sort compares adjacent elements and swaps them if they are in the wrong order, which can lead to many unnecessary comparisons and swaps. In contrast, selection sort selects the minimum element from the unsorted portion of the list and swaps it with the first unsorted element, resulting in fewer comparisons and swaps overall.
```
