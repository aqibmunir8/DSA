```python
import sys
```

```python
# This script tracks the memory size of a list as elements are appended to it.
L = [] # Initialize an empty list
x = 0 # Previous size of the list
t = 1 # Counter for number of elements since last size change
r = 0 # Resize counter
print("append  size  perCellNo  Expand")
for i in range(100):
    if x != sys.getsizeof(L):
        x = sys.getsizeof(L)
        t = 1
        r+=1
    # print(i, sys.getsizeof(L), t, r)
    print(f"{i}\t{sys.getsizeof(L)}\t{t}\t{r}")
    L.append(i)
    t+=1
```
