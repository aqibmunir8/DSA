```python
class Stacks:
    def __init__(self, size):
        self.size = size
        self.stack = [None] * self.size
        self.top = -1
    
    def push(self, value):
        if self.top == self.size - 1:
            raise Exception("Overflow")
        self.top += 1
        self.stack[self.top] = value
    
    def pop(self):
        if self.top == -1:
            raise Exception("Underflow or Empty Stack")
        else: 
            data = self.stack[self.top]
            self.stack[self.top] = None # Clear the position
            self.top -= 1
            return data
    
    def peek(self):
        if self.top == -1:
            raise Exception("Stack is empty")
        return self.stack[self.top]

    def is_empty(self):
        return self.top == -1
    
    def is_full(self):
        return self.top == self.size - 1
    
    def __str__(self):
        if self.top == -1:
            return "Stack is empty"
        elements = [str(self.stack[i]) for i in range(self.top + 1)] 
        print(elements)
        return "Stack elements (bottom to top): " + " -> ".join(elements)
```

```python
L = Stacks(3)
```

```python
print(L)
```
