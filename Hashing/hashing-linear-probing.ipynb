```python
class Dictionary:

    def __init__(self, size):
        self.size = size
        self.slots = [None] * self.size
        self.data = [None] * self.size

    def put(self, key, value):
        hash_value = self.hash_function(key)

        if self.slots[hash_value] == None: # Case 1
            self.slots[hash_value] = key
            self.data[hash_value] = value

        else:

            if self.slots[hash_value] == key: # Case 2
                self.data[hash_value] = value
            else: # Case 3
                new_hash_value = self.rehash(hash_value)
                #        case a                               case b
                while self.slots[new_hash_value] != None and self.slots[new_hash_value] != key:
                    new_hash_value = self.rehash(new_hash_value)

                if self.slots[new_hash_value] == None: # case a
                    self.slots[new_hash_value] = key
                    self.data[new_hash_value] = value
                else:                                 # case b
                    self.data[new_hash_value] = value

    def get(self, key):
        start_position = self.hash_function(key) # Get hash forkey, Example Python=>1
        current_position = start_position        # will use it for loop until find python

        while self.slots[current_position] != None:  # Loop - if-none-encounter- Case-02

            if self.slots[current_position] == key: # Case-00
                return self.data[current_position]
            
            current_position = self.rehash(current_position)  # +=1

            if current_position == start_position: # Case-01
                return "Not Found"

        return "None wala Not Found" # Case-02 Loop Breaks- Not Found - none-encountered

    def __str__(self): # Printing
        result = "{\n"
        for i in range(len(self.slots)):
            if self.slots[i] != None:
                result += f"{self.slots[i]} : {self.data[i]},\n" 
        result+="}"
        return result

    def __getitem__(self,key): # Dictionary["key"]
        return self.get(key)

    def __setitem__(self,key,value): # Dictionary["key"] = "value"
        self.put(key,value)

    def rehash(self, old_hash):
        return (old_hash + 1) % self.size


    def hash_function(self, key):
        return abs(hash(key)) % self.size
    
    def keys(self): # Return all keys
        return [key for key in self.slots if key is not None]
    
    def values(self): # Return all values
        return [self.data[i] for i in range(self.size) if self.slots[i] is not None]
    
    def items(self): # Return all key-value pairs
        return [(self.slots[i], self.data[i]) for i in range(self.size) if self.slots[i] is not None]
    
    def __len__(self): # Return number of key-value pairs
        return len(self.keys())
    
    def __delitem__(self, key): # Delete a key-value pair
        hash_value = self.hash_function(key)
        start_position = hash_value
        current_position = start_position

        while self.slots[current_position] is not None:
            if self.slots[current_position] == key:
                self.slots[current_position] = None
                self.data[current_position] = None
                return
            current_position = self.rehash(current_position)
            if current_position == start_position:
                break
        raise KeyError(f"Key {key} not found.")
    
    def clear(self): # Clear the dictionary
        self.slots = [None] * self.size
        self.data = [None] * self.size
    

    
if __name__ == "__main__": # Example usage
    d = Dictionary(10)
    d["apple"] = 1
    d["banana"] = 2
    d["orange"] = 3
    print(d)
    print(d["banana"])
    del d["banana"]
    print(d)
    print("Keys:", d.keys())
    print("Values:", d.values())
    print("Items:", d.items())
    print("Length:", len(d))
    d.clear()
    print("After clear:", d)
```
